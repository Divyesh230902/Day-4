```python
# from transformers import T5Tokenizer, T5ForConditionalGeneration
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# initialize the model tokenizer
tokenizer = AutoTokenizer.from_pretrained("t5-small")

# initialize the model architecture and weights
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")
```

```python
# article = """Too much noise!! The room they gave me was dirty. The floors looked like they have not been vacuumed. There were people that had the room next to ours that made so much noise that kept us up all night. We informed the front desk, but that didn't seem to stop them from going screaming all night long. Luckily we only stayed for 1 night. 
# Happy New Year at The Hampton! Great Location  We stayed at the Hampton Inn Times Square during our visit to NYC over the New Years holiday. As other reviews have stated, this hotel is a great value for NYC! The hotel is reasonably priced and centrally located in walking distance to Times Square, Rockefeller Center, Broadway, shopping on 5th Ave., Central Park, the Museum of Modern Art, and more. We were definitely pleased that we chose to stay in Midtown Manhattan. While this hotel is not a full service hotel (i.e., there is no bar or restaurant within the hotel), it provides all of the basic accommodations needed. Breakfast is included in the price, which is also a plus. The rooms were nicely apportioned and decorated. The beds were comfortable and we were provided with extra towels without having to ask for them. Front desk staff were helpful in answering questions and assisting with restaurant reservations. We look forward to returning to this hotel in future visits to NYC!"""

article = """After cricketer Kuldeep Yadav got vaccinated at a guest house in Kanpur, the photo of Gujarati folk singer Geeta Rabari getting the shot at her home in Kutch district has stoked a fresh controversy, at a time when struggles for Covid-19 vaccine slots and vaccines has not yet become a thing of the past for common people. The district administration has issued a show-cause notice to the woman health worker, who administered the jab after a complaint was filed with the district administration."""

# article = """It was a Friday night. The sky was showering all over the city. Darkness was spread in each and every corner of the bungalow and nothing could be seen. The sound of lightening in the sky was making it scarier. Suddenly my legs felt some obstacle in my way towards the bedroom. When I tried to see it with the help of little torch, I went numb. It was a dead body of our domestic help, who got brutally murdered."""
```

```python
# encode the text into tensor of integers using the appropriate tokenizer
inputs = tokenizer.encode("summarize: " + article, return_tensors="pt", max_length=512, truncation=True)
```

```python
# generate the summarization output
outputs = model.generate(
    inputs, 
    max_length=450, 
    min_length=10, 
    length_penalty=2.0, 
    num_beams=2, 
    early_stopping=True)
# just for debugging
#print(outputs)
print(tokenizer.decode(outputs[0]))
```
